```python
#Azure Databricks Data Preprocessing

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

print("Spark Version:", spark.version)
```

```python
## Load the dataset
df = spark.read.csv(
    "/Volumes/antonids_databricks_we/nids/nids_volume/balanced_dataset.csv",
    header=True,
    inferSchema=True
)

print("Rows:", df.count())
print("Columns:", len(df.columns))

display(df.limit(10))
```

```python
## Dataset Information

print(df.count())
print(len(df.columns))
```

```python
## Schema Validation

df.printSchema()
```

```python
## Missing Value Analysis

rom pyspark.sql.functions import col, count, when

missing = df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
])

display(missing)
```

```python
# Remove duplicate rows
df = df.dropDuplicates()

# Remove rows with null values
df = df.dropna()

print("Rows after cleaning:", df.count())
```

```python
## Save Processed Dataset

output_path = "/Volumes/antonids_databricks_we/nids/nids_volume/processed_dataset"

df.write.mode("overwrite").parquet(output_path)

print("Processed dataset saved successfully!")
```

```python
## Verify Output

processed_df = spark.read.parquet(output_path)

print("Rows:", processed_df.count())

display(processed_df.limit(10))
```
